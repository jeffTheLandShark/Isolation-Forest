# Load and view dataset

```python
# Common imports and helper function
import os
import pandas as pd
import numpy as np
from scipy import io as sio
```

```python
# 3) vertebral.mat (MAT-file)
data_path = 'vertebral.mat'
mat = sio.loadmat(data_path)
mat.keys()
```

```python
X = mat['X']
y = mat['y']
feature_names = [f'feature_{i}' for i in range(X.shape[1])]
df_vertebral = pd.DataFrame(X, columns=feature_names)
df_vertebral['target'] = y
df_vertebral.describe()
```

```python
df_vertebral['target'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Vertebral Column Target Distribution')
```
